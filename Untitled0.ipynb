```python
%matplotlib inline
from cky import *
from greader import *
from datetime import datetime, timedelta
import sys
from pickle import dumps
import matplotlib.pyplot as plt

```

```python
def show_chart(x, t, m):

    fig, ax1 = plt.subplots()
    fig.set_size_inches(15.5,7.5)
    ax1.set_ylabel('time (s)')
    ax1.plot(x, t, 'r-', label="time(s)")
    ax2 = ax1.twinx()
    ax2.set_ylabel('memory aprox')
    ax2.plot(x, m, 'b--', label="memory aprox")
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()
    
max_length = 100
tries = 1
```

```python
#read the grammar G1 (a simple grammar)
R, NT, T, Gt, GT, valid = parse_cnf(open("G1.g"))

#simple grammar: aaaa...aaab
s = ""
t = []
m = []
x = []
for i in range(1,max_length):
    s = ""
    for j in range(i):
        s = s + ' a'
    s = s + ' b'
    
    #print s
    
    ini = datetime.now()
    for k in range(tries):
        (r, PT) = cky(R, NT, T, Gt, GT, s.split(' '), False, False)
    finish = datetime.now()
    
     
    t.append( (finish-ini).total_seconds()  / tries )
    m.append( len(dumps(PT) ) ) 
    x.append( len(s) )
    #print "%d -> time: %s \tmemory: %s"%(i, t[-1], m[-1])

show_chart(x,t,m)
```

```python
#simple grammar: aaaa...aaabbb.....bbbbb
s = ""
t = []
m = []
x = []
for i in range(1,max_length):
    s = ""
    for j in range(i/2):
        s = s + ' a'
 
    
    for j in range(i-(i/2)):
        s = s + ' b'
        

    #print s
    
    ini = datetime.now()
    for k in range(tries):
        (r, PT) = cky(R, NT, T, Gt, GT, s.split(' '), False, False)
    finish = datetime.now()
    
     
    t.append( (finish-ini).total_seconds() / tries )
    m.append( len(dumps(PT) ) )
    x.append( len(s) )
    #print "%d -> time: %s \tmemory: %s"%(i, t[-1], m[-1])

show_chart(x,t,m)
```
